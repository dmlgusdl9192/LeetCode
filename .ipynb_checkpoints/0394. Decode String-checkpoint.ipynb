```python
from collections import deque
```

```python
# using stack
# s3 = "2[abc]3[cd]ef"
# -> deque([ 2[abc], 3[cd]ef ])
# -> deque([ abc, abc, 3[cd]ef])
# -> ans = abcabc || deque([ 3[cd], ef ])
# -> ans = abcabc || deque([ cd, cd, cd, ef ])
# -> ans = abcabccdcdcdef

# s5 = "3[2[b]a]c"
# -> deque([ 2[b]a, 2[b]a, 2[b]a], c)

# if top item's first letter is a digit, then look for divide the item into two pieces, left and right 
# -> item >>> left: digit & [ ... ], right: rest of item

# if top item's first letter is alphabet, then look for end of string and then append to answer.
# End of string is reached when the next character is empty or digit.
```

```python
def decodeString(s):
    stack = deque([s])
    ans = ""
    while stack:
#         print(stack)
        item = stack.pop()

# if top item's first letter is a digit, then look for divide the item into two pieces, left and right 
# -> item >>> Left: digits & [ left_str ], Right: rest of item

        if item[0].isdigit():
#             print('digit')
            
            # Look for digits part of Left.
            left_index = 0
            digits = ''
            while item[left_index].isdigit():
                digits += item[left_index]
                left_index += 1
            
            # Skip first bracket after digits
            bracket_counter = 1
            left_index += 1
            
            # Look for left_str 
            left_start_index = left_index
            while bracket_counter > 0:
                if item[left_index] == '[':
                    bracket_counter += 1
                elif item[left_index] == ']':
                    bracket_counter -= 1
                left_index += 1
            
            left  = item[left_start_index:left_index-1]
            if left_index < len(item):
                stack.append(item[left_index:])
            for i in range(int(digits)):
                stack.append(left)
            
# if top item's first letter is alphabet, then look for end of characters and then append to answer.
# End of characters is reached when the next index is empty or digit.   
        elif item[0].isalpha():
#             print('alpha')
            
            # Look for characters part of item
            chr_index = 0
            chrs = ''
            while chr_index < len(item) and item[chr_index].isalpha():
                chr_index += 1
            chrs = item[:chr_index]
            
            # append chrs to answer and rest back to stack
            ans += chrs
            if chr_index < len(item):
                stack.append(item[chr_index:])
    return ans
```

```python
s1 = "3[a]2[b]"
s2 = "3[a2[c]]"
s3 = "2[abc]3[cd]ef"
s4 = "abc3[cd]xyz"
s5 = "3[2[b]a]c" # 'bbabbabbac'
l = [s1,s2,s3,s4,s5]

for s in l:
    print(decodeString(s))
    print('-----')
```
