Given an integer n, return the least number of perfect square numbers that sum to n.

A perfect square is an integer that is the square of an integer; in other words, it is the product of some integer with itself. For example, 1, 4, 9, and 16 are perfect squares while 3 and 11 are not.

 

Example 1:

* Input: n = 12
* Output: 3
* Explanation: 12 = 4 + 4 + 4.

Example 2:

* Input: n = 13
* Output: 2
* Explanation: 13 = 4 + 9.
 

Constraints:

* 1 <= n <= 104


```python
# Using BFS (Breadth-first search) method. Each breadth/layer is made up of perfect squares that are 
# smaller than the difference between target n and the running total of each path (sum_so_far)
from collections import deque
def numSquares(n):
    q = deque([0])
    count_sum = 0
    visited = set() # Set of sum_so_far's. A sum_so_far in this set already has a shorter path.
                    # Thus that path can be discarded and not be added to the queue anymore.
    while q:
        count_sum += 1 # count of perf squares summed so far
        for _ in range(len(q)):
            sum_so_far = q.popleft() # sum of perf squares so far
            for x in range(1, n+1):
                if sum_so_far + x**2 > n:
                    break
                elif sum_so_far + x**2 == n:
                    return count_sum
                else:
                    if (sum_so_far + x**2) not in visited:
                        visited.add(sum_so_far + x**2)
                        q.append(sum_so_far + x**2)
                    
        
```

```python
for x in range(1,100):
    print(x, numSquares(x))
```
